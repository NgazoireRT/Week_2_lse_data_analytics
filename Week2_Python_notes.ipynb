# Mapping Values

```python
"""The map() method works like â€˜find and replaceâ€™ in Excel. 
The map() function works by applying a function to every element of a Series (either rows or columns), 
which in turn creates a new value for each element. 
This can either be used to update an existing Series or create a new one. 
For example, you may want to create a new Series that displays the total 
sales prices for a group of items by adding the sales tax to the net listed price."""

## Mapping a Series

# Students tut 2 mark.
students_updated ['Tutorial 2'] = [80, 75, 79, 69, 82, 68, 90, 84, 83, 79]

# Map method on tut 2.
# Create new series.
tut_2 = {80:80, 75:75, 79:76, 69:74, 82:80,
         68:75, 90:80, 84:84, 83:81, 79:80}

# Add new series to current DataFrame.
students_updated['tut_2'] = students_updated ['Tutorial 2'].map(tut_2)

# Reorder columns.
students_updated = students_updated [['Name', 'Age', 'Qualification', 
                                     'Country', 'Test 1', 'Tutorial 1',
                                     'Test 2', 'Tutorial 2', 'tut_2', 
                                     'Test 3', 'Tutorial 3', 'Final mark',
                                     'Pass']]

students_updated
```
